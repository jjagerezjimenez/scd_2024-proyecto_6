```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parámetros
T = 1  # Período de símbolo
num_symbols = 10  # Número de símbolos U_k a considerar en cada dirección
t_start = 0 * T  # Tiempo inicial
t_end = 20 * T     # Tiempo final
t_samples = 1000  # Número de muestras de tiempo
```

```python
# Vector de tiempo
t = np.linspace(t_start, t_end, t_samples)
```

```python
# Generar secuencia de U_k (valores ±1 con probabilidad 1/2)
k_indices = np.arange(0, num_symbols)
U_k = np.random.choice([-1, 1], size=len(k_indices))
```

```python
# Inicializar U(t)
U_t = np.zeros_like(t)

# Calcular U(t) sumando sobre k
for k, Uk in zip(k_indices, U_k):
    U_t += Uk * np.sinc((t - k * T) / T)

# Graficar U(t)
plt.figure(figsize=(12, 6))
plt.plot(t, U_t, label='U(t)')
plt.title('Simulación del Ejemplo 7.2.2: Proceso Aleatorio U(t)')
plt.xlabel('Tiempo t')
plt.ylabel('Amplitud')
plt.grid(True)
plt.legend()
plt.show()
```
